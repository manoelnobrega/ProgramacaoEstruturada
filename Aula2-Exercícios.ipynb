```python
#[c]
%%writefile ex3.c
#include <stdio.h>

int main() {
  float valorT, comiss;
  printf("Informe o valor da transação:\n");
  scanf("%f", valorT);

  if (valorT<=2500) {
    comiss = 30 + (valorT * 0,017);
  }
  else if (valorT>2500 && valorT<=6250) {
    comiss = 56 + (valorT * 0,0066);
  }
  else if (valorT>6250 && valorT<=20000) {
    comiss = 76 + (valorT * 0,0034);
  }
  else if (valorT>20000 && valorT<=50000) {
    comiss = 100 + (valorT * 0,0022);
  }
  else if (valorT>50000 && valorT<=500000) {
    comiss = 155 + (valorT * 0,0011);
  }
  else {
    comiss = 255 + (valorT * 0,0009);
  }

  printf("A comissão será de: %.2f\n", comiss)

}
```

# Exercício 4

```python
#[c]
%%writefile ex4.c
#include <stdio.h>

int main() {
  float cels, fahr, kelv;

  printf("Digite a temperatura em graus Celsius:\n");
  scanf("%f", &cels);

  fahr = (cels*1.8) + 32;
  kelv = cels + 273;

  printf("Em fahrenheit: %.1f °F\nEm kelvin: %.1f K\n", fahr, kelv);
  return 0;
}
```

```python
#[c]
!gcc ex4.c -o output
!./output
```

# Exercício 5

```python
#[c]
%%writefile ex5.c
#include <stdio.h>

int main() {
  int n, numero, soma, i;

  printf("Digite um número:\n");
  scanf("%d", &n);

  printf("Agora digite os números para a soma:\n");
  soma = 0;
  for(i=0, i<n, i++) {
    scanf("%d", &numero);
    soma += numero;
  }

  printf("A soma dos números é igual a %d\n", soma);
  return 0;
}
```

```python
#[c]
!gcc ex5.c -o output
!./output
```

# Exercício 6

```python
#[c]
%%writefile ex6.c
#include <stdio.h>

int main() {
  int dividendo, divisor, cont, resto;

  printf("Entre com o Dividendo:\n");
  scanf("%d", &dividendo);
  printf("Entre com o Divisor:\n");
  scanf("%d", &divisor);

  cont = 0;
  resto = dividendo;
  while(resto >= divisor) {
    resto -= divisor;
    cont++;
  }

  printf("A divisão de %d por %d é igual a %d e tem resto %d\n", dividendo, divisor, cont, resto);
  return 0;
}
```

```python
#[c]
!gcc ex6.c -o output
!./output
```
