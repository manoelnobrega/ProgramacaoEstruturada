# Exercício 1

```python
#[c]
%%writefile ex1.c
#include <stdio.h>

double pot(double a, int b);

int main() {
  double a;
  int b;

  printf("Digite a Base:\n");
  scanf("%lf", &a);
  printf("Digite o Expoente:\n");
  scanf("%d", &b);

  printf("Resultado = %.2lf\n", pot(a, b));
  return 0;
}

double pot(double a, int b) {
  double resul = 1;
  int i;
  int aux = b;

  if(b == 0)
    return resul;

  if(b < 0)
    aux = b*(-1);

  for(i=1; i<=aux; i++) {
    resul = resul * a;
  }

  if(b < 0)
    resul = 1/resul;

  return resul;
}
```

# Exercício 2

```python
#[c]
%%writefile ex2.c
#include <stdio.h>

long int fat(int n);

int main() {
  int n;

  printf("Digite um número para ver seu fatorial:\n");
  scanf("%d", &n);

  printf("Resultado = %ld", fat(n));
  return 0;
}

long int fat(int n) {
  if (n <= 0)
    return 1;
  else
    return n * fat(n-1);
}
```

# Exercício 3

```python
#[c]
%%writefile ex3.c
#include <stdio.h>

int primo(int n, int div);
int verifica_primo_menor(int n);
int verifica_primo_maior(int n);

int main() {
  int n;

  printf("Digite um número: ");
  scanf("%d", &n);

  printf("Os dois números primos mais próximos de %d são %d e %d\n", n, verifica_primo_menor(n), verifica_primo_maior(n));
  return 0;

}

int primo(int n, int div) {
  if (div == 1) {
    return 1;
  }
  else {
    if(n % div == 0)
      return 0;
    else
      return primo(n, div-1);
    }
}

int verifica_primo_menor(int n) {
  int ver, menor;

  for(int i=n; i<=n; i--) {
    ver = primo(i, i-1);
    if (ver == 1) {
      menor = i;
      break;
    }
    else
      continue;
  }

  return menor;
}

int verifica_primo_maior(int n) {
  int ver, maior;

  for(int j=n+1; j>n; j++){
    ver = primo(j, j-1);
    if (ver == 1) {
      maior = j;
      break;
    }
    else
      continue;
  }

  return maior;
}
```

```python
#[c]
!gcc ex3.c -o output
!./output
```

# Exercício 4

```python
#[c]
%%writefile ex4.c
#include <stdio.h>

int fibonacci(int n);

int main() {
  int n;

  printf("Digite um número N para ver a sequência de fibonacci: ");
  scanf("%d", &n);

  printf("o resultado esperado é %d\n", fibonacci(n));

  return 0;
}

int fibonacci(int n) {

  if (n <= 0) {
    return 0;
    }
  else {
    return fibonacci(n-1) + fibonacci(n-2);
  }
}
```
