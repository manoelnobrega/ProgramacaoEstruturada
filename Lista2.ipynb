```python
#[c]
%%writefile ex1.c
#include <stdio.h>

int main() {
  float a, b; // a e b são do tipo float
  float c, *pa, *pb; // c é do tipo float, pa e pb são ponteiros para float

  a = 0.001; // a recebe 0.001
  b = 0.003; // b recebe 0.003
  pa = &a; // pa recebe o endereço de a (1000)
  *pa = 2 * a; // pa agora aponta para o dobro de a
  pb = &b; // pb recebe o endereço de b (1004)
  c = 3 * (*pa + *pb); // c recebe 3 vezes a soma de pa e pb (0.015)

  /*
  (a) pa = 1000
  (b) *pa = 0.002
  (c) pb = 1004
  (d) c = 0.015

  */

  return 0;
}
```

```python
#[c]
%%writefile ex2.c
#include <stdio.h>
#include <stdlib.h>

void incrementaPar(int *v, int tam);

int main() {
  int *v, i;
  int tam;

  printf("Tamanho do vetor: ");
  scanf("%d", &tam);
  v = (int *)malloc(tam * sizeof(int));

  printf("Digite os elementos do vetor:\n");
  for (i = 0; i < tam; i++) {
    scanf("%d", &v[i]);
  }

  incrementaPar(v, tam);

  printf("Vetor com os elementos pares incrementados:\n");
  for (i = 0; i < tam; i++) {
    printf("%d ", v[i]);
  }

  free(v);
  return 0;
}

void incrementaPar(int *v, int tam) {
  for (int i = 0; i < tam; i++) {
    if (i % 2 == 0)
        v[i]++;
  }
  return;
}
```

```python
#[c]
%%writefile ex3.c
#include <stdio.h>
#include <stdlib.h>

int main() {
  int *v, j, i = 0;
  int n;

  printf("Digite um numero (0 para sair): ");
  scanf("%d", &n);

  while (n != 0) {
    v = (int *)realloc(v, (i + 1) * sizeof(int));
    v[i] = n;
    i++;

    printf("Digite um numero (0 para sair): ");
    scanf("%d", &n);
  }

  printf("Vetor: ");
  for(j=0; j < i; j++) {
    printf("%d ", v[j]);
  }

  free(v);
  return 0;
}
```

```python
#[c]
%%writefile ex6.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int n, i;
    double *v;

    printf("Numero de valores no vetor: ");
    scanf("%d", &n);
    while(n < 10) {
        printf("O vetor é muito pequeno. Digite novamente\n");
        scanf("%d", &n);
    }

    v = (double *)malloc(n * sizeof(double));

    for(i=0; i<n; i++) {
        if(i<10) {
            v[i] = rand()%100;
        }
        else {
            printf("Digite um valor para a posição %d do vetor: ", i+1);
            scanf("%lf", &v[i]);
        }
    }

    printf("Os 10 primeiros valores do vetor são:\n");
    for(i=0; i<10; i++) {
        printf("%.1lf ", v[i]);
    }
    printf("\n");

    free(v);
    return 0;
}
```

```python
#[c]
%%writefile ex7.c

#include <stdio.h>
#include <stdlib.h>

int main() {
    int n, i, j;
    int **m;

    printf("Digite a dimensão da matriz: ");
    scanf("%d", &n);

    //alocando a memoria para a matriz
    m = malloc(n * sizeof(int *));
    for(i=0; i<n; i++) {
        m[i] = malloc(n * sizeof(int));
    }

    //preenchendo com zeros usando ponteiros pra endereçar os elementos e não "m[i][j]"
    for(i=0; i<n; i++) {
        for(j=0; j<n; j++) {
            *(*(m+i)+j) = 0;
        }
    }

    //mostrar a matriz na tela
    printf("Matriz %d x %d:\n", n, n);
    for(i=0; i<n; i++) {
        for(j=0; j<n; j++) {
            printf("%d ", m[i][j]);
        }
        printf("\n");
    }

    //preenchendo com numeros de 1 a 10000
    for(i=0; i<n; i++) {
        printf("Digite valores entre 1 e 10000 da linha %d\n", i+1);
        for(j=0; j<n; j++) {
            scanf("%d", (*(m+i)+j)); // no caso do scanf não temos um segundo * pois queremos o endereço de memória e não o conteúdo
        }
    }

    //mostrar a matriz na tela
    printf("Matriz %d x %d:\n", n, n);
    for(i=0; i<n; i++) {
        for(j=0; j<n; j++) {
            printf("%d ", m[i][j]);
        }
        printf("\n");
    }

    //desalocando a memoria
    for(i=0; i<n; i++) {
        free(m[i]);
    }
    free(m);

    return 0;
}
```
