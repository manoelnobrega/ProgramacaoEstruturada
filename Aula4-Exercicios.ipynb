# Exercício 1

```python
#[c]
%%writefile ex1.c
#include <stdio.h>

int main(){
  char st[80];
  char inv[80];
  int fim, i, j = 0;

  printf("Digite uma string para ver sua inversa\n");
  fgets(st, 80, stdin);

  fim = 0;
  while(st[fim] != '\0' && st[fim] != '\n')
    fim++;

  i = fim - 1;
  for(i; i>=0; i--){
    inv[i] = st[j];
    j++;
  }

  inv[fim] = '\0';

  printf("A string inversa é: \n%s\n", inv);
  return 0;

}
```

```python
#[c]
%%writefile ex3.c
#include <stdio.h>

double MediaVetor(double vet[10]);

int main() {
  int i;
  double vet[10];

  for(i=0; i<10; i++) {
    printf("Digite um valor flutuante para a posição %d do vetor: ", i);
    scanf("%lf", &vet[i]);
  }

  printf("A média dos valores desse vetor é : %.2lf\n", MediaVetor(vet));
  return 0;

}

double MediaVetor(double vet[10]) {
  int i;
  double soma = 0;

  for(i=0; i<10; i++){
    soma += vet[i];
  }

  return soma/10;
}
```

```python
#[c]
%%writefile ex4.c
#include <stdio.h>

int EscreverTransposta(int mat[4][4]);

int main(){
  int mat[4][4], i, j;
  int trans[4][4];

  //Lendo a matriz:

  for(i=0; i<4; i++) {
    for(j=0; j<4; j++) {
      printf("Digite o valor da linha %d e coluna %d do vetor: ", i, j);
      scanf("%d", &mat[i][j]);
    }
  }

  //Criando a Transposta:

  for(i=0; i<4; i++) {
    for(j=0; j<4; j++) {
      trans[i][j] = mat[j][i];
    }
  }

  //Printando a Matriz

  printf("Matriz Original:\n");
  for(i=0; i<4; i++) {
    for(j=0; j<4; j++) {
      printf("%d ", mat[i][j]);
    }
    printf("\n");
  }

  //Printando a transposta

  printf("Matriz Transposta:\n");
  for(i=0; i<4; i++) {
    for(j=0; j<4; j++) {
      printf("%d ", trans[i][j]);
    }
    printf("\n");
  }

  return 0;
}
```

```python
#include <stdio.h>
#include <stdlib.h>
#include <stdbool.h>

#define MAX_N 30
#define MAX_M 50

int N, M, K;
int matrix[MAX_N][MAX_M];
int vigiada[MAX_M];
int cameras[MAX_N];
int best_combination[MAX_N];
bool found_solution = false;

void distribuiCameras(int idx, int cam_count) {
    // Se o número de câmeras instaladas é igual a K
    if (cam_count == K) {
        // Verifica se todas as ruas estão vigiadas
        bool all_vigiada = true;
        for (int j = 0; j < M; j++) {
            if (vigiada[j] == 0) {
                all_vigiada = false;
                break;
            }
        }
        if (all_vigiada) {
            found_solution = true;
            for (int i = 0; i < K; i++) {
                best_combination[i] = cameras[i];
            }
        }
        return;
    }

    // Se já atingimos K câmeras ou ultrapassamos o número total de esquinas
    if (cam_count > K || idx >= N) {
        return;
    }

    // Tenta colocar a câmera na esquina atual
    for (int i = idx; i < N; i++) {
        // Incrementa o contador de ruas vigiadas
        for (int j = 0; j < M; j++) {
            if (matrix[i][j] == 1) {
                vigiada[j]++;
            }
        }

        // Adiciona a esquina atual à lista de câmeras
        cameras[cam_count] = i;

        // Chama a função recursivamente para a próxima esquina
        distribuiCameras(i + 1, cam_count + 1);

        // Remove a câmera da esquina atual (backtracking)
        for (int j = 0; j < M; j++) {
            if (matrix[i][j] == 1) {
                vigiada[j]--;
            }
        }
    }

    // Também tenta não colocar a câmera na esquina atual (sem ela)
    distribuiCameras(idx + 1, cam_count);
}

int main() {
    // Leitura da entrada
    scanf("%d %d %d", &N, &M, &K);
    for (int i = 0; i < N; i++) {
        for (int j = 0; j < M; j++) {
            scanf("%d", &matrix[i][j]);
        }
    }

    // Inicializa o vetor de ruas vigiadas
    for (int i = 0; i < M; i++) {
        vigiada[i] = 0;
    }

    // Inicia a distribuição das câmeras
    distribuiCameras(0, 0);

    // Se não encontrou nenhuma combinação válida
    if (!found_solution) {
        printf("-1\n");
    } else {
        // Imprime a melhor combinação encontrada
        for (int i = 0; i < K; i++) {
            printf("%d", best_combination[i]);
            if (i < K - 1) {
                printf(" ");
            }
        }
        printf("\n");
    }

    return 0;
}
```
